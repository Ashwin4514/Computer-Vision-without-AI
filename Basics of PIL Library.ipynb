```python
from PIL import Image 
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Creating Image Object

```python
image1= Image.open("C:/Users/ASHWIN/Desktop/Bumblebee.png") #Specifying the location of the image
```

Viewing Basic properties of the Image 

```python
print(image1.size)
print(image1.format)
print(image1.mode) #There are four channels in this image: Red , Green ,Blue and Alpha.
plt.imshow(image1)
```

Cropping

```python
image2= image1.copy()
image3= image2.crop((350,190,450,320))
plt.imshow(image3)
```

Transposing

```python
image4=image1.transpose(Image.FLIP_LEFT_RIGHT)
image5=image1.transpose(Image.FLIP_TOP_BOTTOM)
image6=image1.transpose(Image.ROTATE_90)
image7=image1.transpose(Image.ROTATE_180)
image8=image1.transpose(Image.ROTATE_270)
image9=image1.transpose(Image.TRANSPOSE)
plt.imshow(image1)
plt.title("Original")
plt.figure(figsize=(10,10))
plt.subplot(2,3,1)
plt.title("Flipping from Left to right")
plt.imshow(image4)
plt.subplot(2,3,2)
plt.title("Flipping TOP TO BOTTOM")
plt.imshow(image5)
plt.subplot(2,3,3)
plt.title("ROTATE:90")
plt.imshow(image6)
plt.subplot(2,3,4)
plt.title("ROTATE:180")
plt.imshow(image7)
plt.subplot(2,3,5)
plt.title("ROTATE:270")
plt.imshow(image8)
plt.subplot(2,3,6)
plt.title("TRANSPOSE")
plt.imshow(image9)
```

RESIZING THE IMAGE USING INTERPOLATION ALGORITHMS

```python
_image= Image.open("C:/Users/ASHWIN/Desktop/thumb-350-618138.jpg")
print(_image.size)
```

```python

image4=_image.resize((175,109),Image.NEAREST) # Introduces Jaggedness
image5=_image.resize((175,109),Image.BOX) 
image6=_image.resize((175,109),Image.BILINEAR)
image7=_image.resize((175,109),Image.HAMMING)
image8=_image.resize((175,109),Image.BICUBIC)
image9=_image.resize((175,109),Image.LANCZOS)
plt.imshow(_image)
plt.title("Original")
plt.figure(figsize=(20,10))
plt.subplot(2,3,1)
plt.title("Nearest")
plt.imshow(image4)
plt.subplot(2,3,2)
plt.title("BOX")
plt.imshow(image5)
plt.subplot(2,3,3)
plt.title("BILINEAR")
plt.imshow(image6)
plt.subplot(2,3,4)
plt.title("HAMMING")
plt.imshow(image7)
plt.subplot(2,3,5)
plt.title("BICUBIC")
plt.imshow(image8)
plt.subplot(2,3,6)
plt.title("LANCZOS")
plt.imshow(image9)
```

TEXT WATERMARK

```python
from PIL import ImageDraw
from PIL import ImageFont
_image2= _image.copy()
draw= ImageDraw.Draw(_image2)
font= ImageFont.truetype("arial.ttf",15)
draw.text((0,0),"Its Showtime",(200,40,30),font=font)
plt.imshow(_image2)
```

IMAGE WATERMARK

```python
size= (70,70) 
_image3= _image2.copy()
_image3.thumbnail(size)

copied_image= _image2.copy()
copied_image.paste(_image3,(300,0))
plt.imshow(copied_image)
```

Enhancement

```python
from  PIL import ImageEnhance as ImE
image4=ImE.Color(_image2).enhance(1.5)
image5=ImE.Brightness(_image2).enhance(1.5)
image6=ImE.Contrast(_image2).enhance(1.5)
image7=ImE.Sharpness(_image2).enhance(1.5)

plt.imshow(_image2)
plt.title("Original")
plt.figure(figsize=(15,15))
plt.subplot(1,4,1)
plt.title("Color")
plt.imshow(image4)
plt.subplot(1,4,2)
plt.title("Brightness")
plt.imshow(image5)
plt.subplot(1,4,3)
plt.title("Contrast")
plt.imshow(image6)
plt.subplot(1,4,4)
plt.title("Sharpness")
plt.imshow(image7)
```

Alpha Blending

```python
#It is a method known to blend two images together.
_imageback= Image.open("C:/Users/ASHWIN/Desktop/destruction.png")
_imageback= _imageback.resize(image1.size)
Blended_image= Image.blend(image1,_imageback,0.67)
plt.figure(figsize=(10,10))
plt.subplot(1,3,1)
plt.imshow(image1)
plt.subplot(1,3,2)
plt.imshow(_imageback)
plt.subplot(1,3,3)
plt.imshow(Blended_image)
```
